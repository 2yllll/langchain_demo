# Transformer Pipeline

将 预处理，后处理，连接起来，形成 完成的数据流

文本 经过 Pipeline，三步：

+ 预处理: 文本 --> 模型可理解格式
+ 传入 模型
+ 模型输出 被 后处理 赋予 意义

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

classifier("I've been waiting for a HuggingFace course my whole life.")
# output: [{'label': 'POSITIVE', 'score': 0.9598047137260437}]

classifier(["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"])
# output: [{'label': 'POSITIVE', 'score': 0.9598047137260437},  {'label': 'NEGATIVE', 'score': 0.9994558095932007}]
```

当前库 可用的 `Pipeline`:

+ Zero-shot-classification 文本分类
+ Text-generation 根据提示，补全文本
+ Feature-extraction 获取文本 的 向量表征
+ Fill-mask 填充给定文本中的空白
+ Ner (名称实体识别) 分辨输入中的哪子串分别对应于哪些实体（如人、地点、组织）
+ Question-answering 从文本中找出给定问题的答案。
+ Sentiment-analysis 情感分析：正面，负面 等
+ Summarization 摘要
+ Translation 翻译
